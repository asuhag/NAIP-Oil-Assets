```
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("gdrive/My Drive/EDF Work/New_6/Images/models/detection_model-ex-040--loss-0029.999.h5") 
detector.setJsonPath("gdrive/My Drive/EDF Work/New_6/Images/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="gdrive/My Drive/EDF Work/New_6/Images/Copy of 1m_scale_NAIP_compress3.png", output_image_path="gdrive/My Drive/EDF Work/New_6/Images/detected-Copy of 1m_scale_NAIP_compress3.png")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```

```
# initialize dictionaries for NAIP image corners
#   y1,x1 are lat,long of upper left corner of image
#   y2,x2 are lat,long of upper right corner of image
#   y3,x3 are lat,long of lower right corner of image
#   y4,x4 are lat,long of lower left corner of image
x1=-102.71661978406141
y1=32.039381866754546
x2=-102.61087637585828
y2=32.039672895380384
x3=-102.61053305310438
y3=31.963683047732065
x4=-102.71730642956922
y4=31.96455685197785

# loop over all the detected wind turbines and calculate the 
# latitude and longitude coordinates based on its location in
# full NAIP images with respect to the image corners
latlongWTs=[]
centroid = []

delta_lat = 32.039672895380384 - 31.96455685197785 

delta_long = -102.61053305310438 - (-102.71661978406141)

width = 6549
height = 2183

dlong_w = delta_long/width
dlat_h = delta_lat/height

for i in detections:
  print(i['box_points'][0])
  lat_left = x1+i['box_points'][0]*dlong_w
  long_left = y1+ i['box_points'][1]*dlat_h
  lat_right = lat_left + (i['box_points'][0] - i['box_points'][2])*dlong_w
  long_right = long_left + (i['box_points'][0] - i['box_points'][3])*dlat_h
  cent_lat = (lat_left+lat_right)/2
  cent_long = (long_left+long_right)/2
  latlongWTs.append((lat_left,long_left,lat_right,long_right))
  centroid.append((cent_long,cent_lat))


# print out latitude and longitude coordinates for verification or plotting
print(latlongWTs)
```
